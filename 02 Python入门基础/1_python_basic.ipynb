### 1. 变量以及基本的数据结构


```python
# ==================================================
# 第一章：冒险者的状态
# 教学点：
# 1. 变量的定义
# 2. 基本的数据结构：
# 字符串：存储文本信息
# 整数：存储整数值
# 浮点数：存储带小数的数值
# 布尔值：存储真/假状态
# ==================================================


# 定义变量：变量就像贴了标签的储物箱，可以在内存中存储数据
name = "小明"          # 字符串：存储文本信息
days = 3              # 整数：存储整数值
temp = 36.8           # 浮点数：存储带小数的数值
hungry = True         # 布尔值：存储真/假状态
```

```python
# 查看变量
print(name)
print(days)
print(temp)
print(hungry)
```

```python
# 使用type()函数检查数据类型
print(type(name))
print(type(days))
print(type(temp))
print(type(hungry))
```

```python
# 转换为字符串
print(str(days))         # 整数转换为字符串
print(str(temp))         # 浮点数转换为字符串
print(str(hungry))       # 布尔值转换为字符串

# 转换为字符串查看转换后的类型
print(type(str(days)))
print(type(str(temp)))
print(type(str(hungry)))
```

```python
# 转换为整数
print(int(temp))         # 浮点数转换为整数（小数部分丢失）
# 转换为浮点数  
print(float(days))       # 整数转换为浮点数

print(type(int(temp)))
```

```python
# 转换为布尔值
print(bool(0))           # 零转换为False
print(bool(1))           # 零转换为True
print(bool(2))           # 零转换为True
print(bool(""))          # 空字符串转换为False
print(bool(days))        # 非零整数转换为True
print(bool("你好"))       # 非空字符串转换为True
```

```python
# 转换后的值赋给新的变量
print(float(days)) 
print(days)
# 转换后的值赋给新的变量
days_f = float(days)
print(days_f)
# 直接覆盖原变量
days = float(days)
print(days)
```

```python
# 字符串索引和切片
fishing_log = "小明今天去捕鱼，收获颇丰，需要详细记录"

print("第一个字符:", fishing_log[0]) #  索引从0开始, 也就是第一个字符是0，第二个字符是1
print("最后一个字符:", fishing_log[-1]) # 负数索引从末尾开始，-1是最后一个字符
print("切片(1:3):", fishing_log[1:3]) #  切片从索引1开始到索引3结束，但不包括索引3，也就是第2个字符到第3个字符
print("切片(:2):", fishing_log[:2]) #  从头开始到索引2结束，但不包括索引2，也就是第1个字符到第2个字符
print("切片(1:):", fishing_log[1:]) # 从索引1开始到结尾，也就是从第2个字符到最后
print("切片(1:8:2):", fishing_log[1:8:3]) # 从索引1开始到索引8结束，但不包括索引8，步长为2，也就是每隔一个字符取一个
print("切片(::-1):", fishing_log[::-1])  # 反转字符串，应该适合古人阅读吧
```

```python
# 2.2.1. 数字运算 - 捕鱼数量计算
# 基本算术运算

print("1. 数字运算 - 捕鱼数量计算")
print("小明今天在不同捕鱼点的收获：")
# 创建变量记录捕鱼数量
morning_fish = 8
afternoon_fish = 12
evening_fish = 5
# 数字变量查看
print(morning_fish) # 早上捕鱼数量
print(afternoon_fish) # 下午捕鱼数量
print(evening_fish) # 晚上捕鱼数量

# 转换为字符串输出
print("早上捕鱼数量: " + str(morning_fish))
print("下午捕鱼数量: " + str(afternoon_fish))
print("晚上捕鱼数量: " + str(evening_fish))
```

```python
# 加法运算
total_fish = morning_fish + afternoon_fish + evening_fish

# 输出数值
print(total_fish)
# 转换为字符串输出
print("总捕鱼数量: " + str(total_fish))
```

```python
# 减法运算
eat_fish = 15
remaining_fish = total_fish - eat_fish
print(remaining_fish)
print("剩余鱼数量: " + str(total_fish) + " - " + str(eat_fish) + " = " + str(remaining_fish))

```

```python
# 乘法运算
fish_weight = 2.5  # 每条鱼平均重量
total_weight = total_fish * fish_weight 
print(total_weight) 

print("总重量: " + str(total_fish) + " * " + str(fish_weight) + " = " + str(total_weight) + "公斤")
```

```python
# 除法运算

fishing_hours = 6 #捕鱼总时长
fish_per_hour = total_fish / fishing_hours # 每小时捕鱼数

print(fish_per_hour)
print("每小时捕鱼数: " + str(total_fish) + " / " + str(fishing_hours) + " = " + str(fish_per_hour)) # 除法运算
```

```python
# 整除运算
fishermen = 3 # 渔夫数量
fish_per_person = total_fish // fishermen # 每人分得鱼数
print(fish_per_person)  # 每个人分得鱼数

print("每人分得鱼数: " + str(total_fish) + " // " + str(fishermen) + " = " + str(fish_per_person))
```

```python
# 取余运算
remainder_fish = total_fish % fishermen  # 分配后剩余鱼数
print(remainder_fish)  # 分配后剩余鱼数 
print("分配后剩余鱼数: " + str(total_fish) + " % " + str(fishermen) + " = " + str(remainder_fish))  
```

```python
# 幂运算
fish_growth = 2 # 鱼群每天增长率
days = 5 # 天数
projected_fish = fish_growth ** days # 5天后鱼群数量
print(projected_fish) # 5天后鱼群数量   
 
print("5天后鱼群数量: " + str(fish_growth) + " ** " + str(days) + " = " + str(projected_fish))

print()
```

```python
# 2.2.4. 逻辑运算 - 捕鱼条件判断
print("判断是否满足捕鱼条件：")

# 捕鱼条件变量
has_fishing_rod = True # 是否有鱼竿
has_bait = True # 是否有鱼饵
has_boat = False    # 是否有船
weather_good = True # 天气是否良好
tide_low = False # 是否是退潮
fish_count = 8 # 捕鱼数量

print("有鱼竿: " + str(has_fishing_rod))
print("有鱼饵: " + str(has_bait))
print("有船: " + str(has_boat))
print("天气好: " + str(weather_good))
print("低潮: " + str(tide_low))
print("鱼数: " + str(fish_count))
```

```python

# and 运算 - 所有条件都为真时返回真
print("有鱼竿 and 有鱼饵: " + str(has_fishing_rod and has_bait))  
print("有鱼竿 and 有船: " + str(has_fishing_rod and has_boat))
print("天气好 and 低潮: " + str(weather_good and tide_low))


```

```python
# or 运算 - 至少一个条件为真时返回真
print("有鱼竿 or 有船: " + str(has_fishing_rod or has_boat))  
print("有船 or 低潮: " + str(has_boat or tide_low))  
print("天气好 or 低潮: " + str(weather_good or tide_low))
```

```python
# not 运算 - 取反
print("not 有船: " + str(not has_boat))
print("not 天气好: " + str(not weather_good))
print("not 低潮: " + str(not tide_low))
```

```python
# 复杂逻辑运算
print("有鱼竿 and 有鱼饵 and 天气好: " + str(has_fishing_rod and has_bait and weather_good))
print("有船 or (有鱼竿 and 鱼数>5): " + str(has_boat or (has_fishing_rod and fish_count > 5)))
print("(天气好 and not 低潮) or 有船: " + str((weather_good and not tide_low) or has_boat))
```

```python
# ==================================================
# 第四章：食物分级管理
# 在这一章中，你将学习字典(Dictionary)数据结构
# 教学点：
# 1. 字典概念 - 键值对集合，用于存储关联数据
# 2. 字典的创建和基本操作
# 3. 字典的增删改查操作
# 4. 获取字典的key列表和判断键是否存在
 
# 常用操作：
# - 增加: dict[新键] = 值
# - 删除: dict.pop(键)
# - 修改: dict[已有键] = 新值
# - 查询: dict[键] 或 dict.get(键, 默认值)
# - 获取所有键: dict.keys()
# - 判断键是否存在: 键 in dict
# - 获取所有值: dict.values()
# - 获取所有键值对: dict.items()
# - 获取长度: len(dict)
# ==================================================

print("\n" + "="*50)
print("你在岛上发现了各种可食用的生物和果实")
print("需要对这些食物进行分级管理，方便查找和使用...")

print("""
==========================================
          食物分级系统说明
==========================================

你将使用字典来管理食物信息：
- 键(Key)：食物名称
- 值(Value)：食物的重要程度等级

等级说明：
1 - 生存必需品
2 - 重要食物  
3 - 普通食物
4 - 应急食物
==========================================
""")
```

```python
# 1. 创建字典 - 使用花括号{}和键值对
# 字典是无序的键值对集合(展示顺序由元素的哈希值和内部存储机制决定,相同的集合在不同运行时可能以不同顺序展示)
# 字典使用花括号 {} 创建，元素格式为 key: value
# 字典通过键来快速访问对应的值
# 字典使用花括号{}创建，每个键值对用冒号:分隔，不同键值对用逗号,分隔
food_grades = {
    "椰子": 1,      # 键:"椰子" 对应 值:1
    "淡水鱼": 2,    # 键:"淡水鱼" 对应 值:2
    "海水鱼": 2,    # 键:"海水鱼" 对应 值:2
    "野果": 3,      # 键:"野果" 对应 值:3
    "蘑菇": 4       # 键:"蘑菇" 对应 值:4
}
print(food_grades)  # 打印整个字典

```

```python
# 2. 字典的增加操作 - 添加新的键值对 
# 使用 字典名[新键] = 值 的方式添加新元素
food_grades["海带"] = 3        # 添加新食物"海带"，等级为3
food_grades["海鸟蛋"] = 1      # 添加新食物"海鸟蛋"，等级为1
print(food_grades)  # 打印添加新元素后的字典
```

```python
# 3. 字典的修改操作 - 修改已有键的值
# 使用 字典名[已有键] = 新值 的方式修改元素
food_grades["野果"] = 2        # 将"野果"的等级从3修改为2
print(food_grades)  # 打印修改后的字典
```

```python
# 4. 字典的删除操作 - 删除指定的键值对
# 使用 pop() 方法删除指定键的元素，并返回对应的值
food_grades.pop("蘑菇")  # 删除键为"蘑菇"的元素

```

```python
print(food_grades)  # 打印删除后的字典
```

```python
# 5. 获取字典的key列表
# 使用 keys() 方法获取字典中所有的键
all_keys = food_grades.keys()  # 获取所有食物的名称（键）
print("所有食物名称:", all_keys) # 将键视图转换为列表显示
type(all_keys)
```

```python
# 6. 判断键是否存在
# 使用 in 关键字检查某个键是否存在于字典中
print("是否存在椰子:", "椰子" in food_grades)    # 检查"椰子"是否存在
print("是否存在老虎:", "老虎" in food_grades)    # 检查"老虎"是否存在
```

```python
# 7. 字典的查询操作 - 访问字典中的值
# 使用 字典名[键] 的方式访问值
print("椰子的等级:", food_grades["椰子"])        # 直接访问，如果键不存在会报错
print("淡水鱼的等级:", food_grades.get("淡水鱼")) # 使用get方法访问，更安全
print("淡水鱼的等级:", food_grades.get("老虎")) # 使用get方法访问，更安全
print("未知食物的等级:", food_grades.get("老虎", "未收录")) # 使用get方法访问不存在的键，可以设置默认返回值
```

```python
# 8. 获取键值对
# 获取所有值
all_values = food_grades.values()
print(all_values) # 打印所有值视图

# 获取所有键值对的列表
all_items = food_grades.items()
print(all_items) # 打印键值对视图

print("所有键值对:", list(all_items)) 
# 获取字典长度（键值对数量）
type(all_items)
print("食物种类数量:", len(food_grades))
type(all_items)
```

```python
# 2. if-else语句
print("检查食物库存：")

food_supply = 3
print("食物库存:", food_supply, "天")

if food_supply > 5:
    print("食物充足，可以安心探索")
else:
    print("食物不足，需要尽快寻找食物")
```

```python
# 3. if-elif-else语句
print("根据潮汐决定捕鱼策略：")

tide = "低潮"
print("当前潮汐:", tide)

if tide == "高潮":
    print("高潮时期，适合深海捕鱼")
elif tide == "低潮":
    print("低潮时期，适合岸边捕鱼")
elif tide == "平潮":
    print("平潮时期，适合中等深度捕鱼")
else:
    print("未知潮汐状态，谨慎行动")

print()
```

```python
# 4. 嵌套条件语句 

print("综合判断捕鱼条件：")
weather = "晴朗"
wind_speed = 15
has_equipment = True 

print("天气:", weather)
print("风速:", wind_speed, "km/h")
print("有渔具:", has_equipment)

if weather == "晴朗":
    if wind_speed < 20:
        if has_equipment:
            print("所有条件满足，可以出海捕鱼")
        else:
            print("天气和风速合适，但没有渔具，不出海")
    else:
        print("风速太大，不适合出海")
else:
    print("天气不好，不适合出海")

print()
```

```python
# 5. 逻辑运算符在条件中的使用
print("使用逻辑运算符简化条件判断：")
time = "早晨"
energy_level = 80
has_bait = True
print("时间:", time)
print("精力值:", energy_level)
print("有鱼饵:", has_bait)

# 组合使用逻辑运算符
if (time == "早晨" or time == "傍晚") and energy_level > 60 and has_bait:     #  # 早晨或傍晚且精力值大于60且有鱼饵
    print("完美条件！立即出发捕鱼")
else:
    print("条件不理想，暂时不出海") 
```

```python
# 6. 条件表达式（三元运算符）
# 三元运算符是Python中一种简洁的条件表达式写法，格式为：值1 if 条件 else 值2，可以理解为“如果条件为真，则取值1，否则取值2”
# 适用于简单的条件赋值
is_raining = True
action = "室内活动" if is_raining else "户外活动" 
print("是否下雨:", is_raining)
print("行动建议:", action)
```

```python
# 小明的15天捕鱼记录
fishing_data = [2, 5, 7, 5, 13, 10, 8, 6, 4, 12, 9, 11, 3, 14, 9]
print("15天捕鱼记录:", fishing_data)

#python本身就自带了一些常用的函数，比如sum()函数可以计算总和，len()函数可以计算长度，在这里就可以用来计算总捕鱼量和天数
#计算总捕鱼量和天数
fishing_data = [2,5,7,5,13,10,8,6,4,12,9,11,3,14,9]  #每天捕获的鱼的数量，一共15天
sum(fishing_data)  #计算总共捕获的鱼的数量，自带的sum()函数
print("总捕鱼量:", sum(fishing_data))  #打印总捕鱼量
len(fishing_data)  #计算天数，自带的len()函数
print("天数:", len(fishing_data))  #打印天数


#但是计算中位数就没有自带的函数，如果不借助外力，就需要自己编写函数
# 1. 自己编写中位数函数的问题
def my_median(data):
    """自己编写的中位数计算函数"""
    # 先排序数据
    sorted_data = sorted(data)
    n = len(sorted_data)
    # 计算中位数
    if n % 2 == 1:
        # 奇数个数据，取中间值
        return sorted_data[n // 2]
    else:
        # 偶数个数据，取中间两个值的平均
        return (sorted_data[n // 2 - 1] + sorted_data[n // 2]) / 2
# 使用自定义函数
custom_median = my_median(fishing_data)
print(f"自定义函数计算的中位数: {custom_median}")
```
