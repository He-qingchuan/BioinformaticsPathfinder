### 1.1 plot()函数

```python
# 这段代码的含义是生成一个包含300个元素的数组，这些元素均匀分布在-10到10之间
x=np.linspace(-10,10,300)
# 这段代码的含义是计算数组x中每个元素的正弦值
y=np.sin(x)
```

## 三、matplotlib绘制高级图形及方法

### 3.1 通过官网实例绘制各种饼图，并学习使用plt.subplots()绘图

#### 标记切片

绘制动物的饼图并为切片添加标签。要添加标签，请将标签列表传给 *labels* 参数

```python
import matplotlib.pyplot as plt

labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'  # 每个部分的标签,分别表示青蛙、猪、狗和木头
sizes = [15, 30, 45, 10]

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels)
```

饼图的每个切片都是 `.patches.Wedge` 对象；因此，除了此处展示的自定义外，每个楔形区域都可以通过 *wedgeprops* 参数进行自定义，
#### 自动标注切片

向 *autopct* 传入函数或格式字符串以对切片进行标注。



```python
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%')
```

默认情况下，标签值来自切片的百分比大小。

#### 切片颜色

向 *colors* 传入颜色列表以设置各切片的颜色。



```python
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels,
       colors=['olivedrab', 'rosybrown', 'gray', 'saddlebrown'])
```

#### 填充图案切片

向 *hatch* 传入填充图案列表以设置每个切片的图案。



```python
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, hatch=['**O', 'oO', 'O.O', '.||.'])
```

#### 交换标签和百分比文本的位置

使用 *labeldistance* 和 *pctdistance* 参数分别定位 *labels* 和 *autopct* 文本。



```python
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%',
       pctdistance=1.25, labeldistance=.6)
```

*labeldistance* 和 *pctdistance* 是半径的比率；因此它们在 ``0``（饼图中心）与 ``1``（饼图边缘）之间变化，也可以设置大于 ``1`` 将文本放在饼图外部。

#### 分离、阴影与旋转切片

除了基本饼图外，本演示展示了几个可选功能：

* 使用 *explode* 偏移某个切片
* 使用 *shadow* 添加投影
* 使用 *startangle* 自定义起始角度

此示例对切片进行排序、分离（explode）并旋转它们。



```python
explode = (0, 0.1, 0, 0)  # 仅将第2个切片（即 'Hogs'）分离

fig, ax = plt.subplots()
ax.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
       shadow=True, startangle=90)
plt.show()
```

默认的 *startangle* 为 0，这会使第一个切片（"Frogs"）从正 x 轴开始。此示例设置 ``startangle = 90``，使所有切片逆时针旋转 90 度，并使 Frogs 切片从正 y 轴开始。

#### 控制大小

通过更改 *radius* 参数，通常也调整文本大小以获得更好的视觉效果，饼图可以被缩放。



```python
fig, ax = plt.subplots()

ax.pie(sizes, labels=labels, autopct='%.0f%%',
       textprops={'size': 'smaller'}, radius=0.8)
plt.show()
```

#### 修改阴影

参数 *shadow* 可选地接受一个字典，用于传递给 `.Shadow` 补丁的参数。可用于修改默认阴影。



```python
fig, ax = plt.subplots()
ax.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
       shadow={'ox': -0.05, 'edgecolor': 'b', 'shade': 0.5}, startangle=90) #shadow 阴影,ox 阴影的水平偏移量,edgecolor 阴影的边缘颜色,shade 阴影的透明度
plt.show()
```

### 3.2 根据官网示例修改代码适应自己数据

#### 示例代码演示

```python
import matplotlib.pyplot as plt
import numpy as np

import matplotlib
import matplotlib as mpl


vegetables = ["cucumber", "tomato", "lettuce", "asparagus",
              "potato", "wheat", "barley"]                      #  蔬菜：黄瓜、番茄、生菜、芦笋、马铃薯、小麦、大麦
farmers = ["Farmer Joe", "Upland Bros.", "Smith Gardening",
           "Agrifun", "Organiculture", "BioGoods Ltd.", "Cornylee Corp."]         #  农场主： Joe、Upland Bros.、Smith Gardening、Agrifun、Organiculture、BioGoods Ltd.、Cornylee Corp.

harvest = np.array([[0.8, 2.4, 2.5, 3.9, 0.0, 4.0, 0.0],
                    [2.4, 0.0, 4.0, 1.0, 2.7, 0.0, 0.0],
                    [1.1, 2.4, 0.8, 4.3, 1.9, 4.4, 0.0],
                    [0.6, 0.0, 0.3, 0.0, 3.1, 0.0, 0.0],
                    [0.7, 1.7, 0.6, 2.6, 2.2, 6.2, 0.0],
                    [1.3, 1.2, 0.0, 0.0, 0.0, 3.2, 5.1],
                    [0.1, 2.0, 0.0, 1.4, 0.0, 1.9, 6.3]])


fig, ax = plt.subplots()    #  figsize=(10, 12) 表示图表的宽度为10英寸，高度为12英寸
im = ax.imshow(harvest)




# Show all ticks and label them with the respective list entries，翻译为：显示所有刻度并将其与相应的列表条目标签
ax.set_xticks(range(len(farmers)), labels=farmers,
              rotation=45, ha="right", rotation_mode="anchor") #  rotation_mode="anchor" 表示刻度标签的旋转模式为锚点旋转
ax.set_yticks(range(len(vegetables)), labels=vegetables)

# Loop over data dimensions and create text annotations. ，翻译为：遍历数据维度并创建文本注释
for i in range(len(vegetables)):
    for j in range(len(farmers)):
        text = ax.text(j, i, harvest[i, j],
                       ha="center", va="center", color="w")

ax.set_title("Harvest of local farmers (in tons/year)")
fig.tight_layout()
plt.show()
```
